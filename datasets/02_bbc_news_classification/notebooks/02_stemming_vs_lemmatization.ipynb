# 02. Stemming vs Lemmatization

**Goal:** Compare the effects of PorterStemmer and WordNetLemmatizer on vocabulary reduction.

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import sys
sys.path.append('../../../shared')
from text_utils import apply_stemming, apply_lemmatization
```

## 1. Load Cleaned Data

```
df = pd.read_csv('../data/processed/cleaned_bbc.csv')
# We'll test on a subset to save time if needed, but 2225 rows is small enough.
df = df.dropna(subset=['cleaned_text'])
```

> **📌 Decision Note — Why Morphological Reduction?**
>
> **Chosen approach:** Compare Stemming vs Lemmatization
>
> **Why this works:** Allows us to see which method reduces vocabulary more effectively without losing meaning.
>
> **Alternatives we could have used:**
> | Option | Pros | Cons |
> |--------|------|------|
> | Stemming (Porter) | Fast, aggressive reduction | Can create non-words (e.g., 'organization' -> 'organ'), causing semantic collision. |
> | Lemmatization (WordNet) | Creates real words, preserves meaning better | Slower, requires POS tags to be truly effective. |
>
> **Why we chose this over alternatives:** Comparing them empirically is the best way to choose for a specific dataset.

## 2. Apply Stemming

```
df['stemmed'] = df['cleaned_text'].apply(apply_stemming)
```

## 3. Apply Lemmatization

```
df['lemmatized'] = df['cleaned_text'].apply(apply_lemmatization)
```

## 4. Compare Vocabulary Sizes

```
vec_clean = CountVectorizer()
vec_clean.fit(df['cleaned_text'])
print('Original Vocab:', len(vec_clean.vocabulary_))

vec_stem = CountVectorizer()
vec_stem.fit(df['stemmed'])
print('Stemmed Vocab:', len(vec_stem.vocabulary_))

vec_lem = CountVectorizer()
vec_lem.fit(df['lemmatized'])
print('Lemmatized Vocab:', len(vec_lem.vocabulary_))
```

```
df.to_csv('../data/processed/morph_bbc.csv', index=False)
print('Saved morphologically processed data.')
```
