# 03. Text Vectorization

**Goal:** Apply TF-IDF and tune `max_features` and `min_df`.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import joblib
```

## 1. Load Lemmatized Data

```python
df = pd.read_csv('../data/processed/morph_bbc.csv')
df = df.dropna(subset=['lemmatized'])
X_train, X_test, y_train, y_test = train_test_split(df['lemmatized'], df['category'], test_size=0.2, random_state=42)
```

> **📌 Decision Note — Why Vectorization Tuning?**
>
> **Chosen approach:** TF-IDF with min_df=5 and max_features=5000
>
> **Why this works:** Removes ultra-rare words (typos, obscure names) and limits memory footprint while capturing the most important signals.
>
> **Alternatives we could have used:**
> | Option | Pros | Cons |
> |--------|------|------|
> | No limits (default) | Captures all information | Creates massive, highly sparse matrix. Prone to overfitting on rare words. |
> | High min_df (e.g. 50) | Very dense matrix | Might filter out words that strongly indicate a specific, rare topic. |
>
> **Why we chose this over alternatives:** A `min_df` of 5 drops words appearing in less than 5 documents, effectively pruning typos and rare entities.

## 2. TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer(stop_words='english', min_df=5, max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
print('Final Vocabulary Size:', len(vectorizer.vocabulary_))
```

```python
joblib.dump((X_train_tfidf, X_test_tfidf, y_train, y_test, vectorizer), '../data/processed/tfidf_bbc.pkl')
print('Saved vectorized data.')
```
