# 02. N-Gram Vectorization

**Goal:** Explore how Bigrams (N-grams where N=2) capture negations, and how this impacts vocabulary size.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import joblib
```

## 1. Load Data

```python
df = pd.read_csv('../data/processed/cleaned_imdb.csv')
df = df.dropna(subset=['cleaned_text'])
X_train, X_test, y_train, y_test = train_test_split(df['cleaned_text'], df['label'], test_size=0.2, random_state=42)
```

> **📌 Decision Note — Why N-Gram Selection?**
>
> **Chosen approach:** Use CountVectorizer with ngram_range=(1, 2) (Unigrams + Bigrams)
>
> **Why this works:** Captures two-word combos like 'not good' or 'very bad', drastically improving sentiment polarity detection.
>
> **Alternatives we could have used:**
> | Option | Pros | Cons |
> |--------|------|------|
> | Unigrams only (1, 1) | Small, fast matrix | Misses all context and negations. |
> | Trigrams (1, 3) | Even more context | Causes exponential explosion in feature size (millions of columns), leading to massive sparsity and memory issues. |
>
> **Why we chose this over alternatives:** Bigrams offer the best tradeoff between capturing crucial local context (negations) and keeping feature size manageable.

## 2. Unigram vs Bigram Matrix Size Comparison

```python
vec_uni = CountVectorizer(ngram_range=(1,1), min_df=5)
vec_uni.fit(X_train)
print('Unigram Vocabulary Size:', len(vec_uni.vocabulary_))

vec_bi = CountVectorizer(ngram_range=(1,2), min_df=5)
vec_bi.fit(X_train)
print('Unigram+Bigram Vocabulary Size:', len(vec_bi.vocabulary_))
```

## 3. Save Bigram Vectorized Data

```python
X_train_vec = vec_bi.transform(X_train)
X_test_vec = vec_bi.transform(X_test)
joblib.dump((X_train_vec, X_test_vec, y_train, y_test, vec_bi), '../data/processed/bigram_data.pkl')
print('Saved Bigram matrices.')
```
