# 04. Model Comparison

**Goal:** Compare NB against Logistic Regression and SVM to find the best model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import joblib
import sys
sys.path.append('../../../shared')
from evaluation_utils import evaluate_classification
```

## 1. Load Data (Using TF-IDF)

```python
X_train, X_test, y_train, y_test, vectorizer = joblib.load('../data/processed/tfidf_data.pkl')
```

## 2. Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
evaluate_classification(y_test, y_pred_lr, labels=['ham', 'spam'])
```

## 3. Support Vector Machine (Linear Kernel)

```python
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
evaluate_classification(y_test, y_pred_svc, labels=['ham', 'spam'])
```

## Key Takeaways
- [x] SVM (Linear) often provides the absolute best separation for high-dimensional sparse text data.
- [x] Logistic Regression is a strong, well-calibrated alternative.
- [x] For spam detection, we care most about minimizing False Positives (we don't want a legitimate text marked as spam). SVM typically excels here.
